```python
#Q)1페이지에 있는 100개의 2024.08.28의 100개 공시 정보 출력/DF
#  : 시간, 시장종류,회사명,회사코드,공시명,공시코드,요청일
#Q)2024.08.28의 모든 공시 정보를 셀레니움을 이용해서 1개의 DF에 담아보세요!
#            : 어제 했던 대로 url을 직접 page별로 만들어서 쏘지 말고,,
#            "밑에 있는 버튼 클릭으로 이동해보세요!!"
```

```python
# 1. 시도 1) 숫자만들어서 돌리겠습니다..
result = [] # --> 개별 공시에 대한 정보들을 담을 리스트!!!
#             --> 개별 공시에 대해서 리스트로 처리하실지, dict 하실지는 선택!!

# 해당하는 공시가 몇 개인 확인!!!
tot_num = len(driver.find_elements(By.TAG_NAME, "tr"))
#tot_num
for i in range(1,tot_num): # 1~~~100
    # 공시 제출 시간
    time_path = f'//*[@id="listContents"]/div[2]/table/tbody/tr[{i}]/td[1]'
    t_data = driver.find_element( By.XPATH,time_path ).text
    print(t_data)
    # 공시 회사 이름, 시장, 고유값.
    co_path = f'//*[@id="listContents"]/div[2]/table/tbody/tr[{i}]/td[2]/span/span'
    co_cat = driver.find_element(By.XPATH, co_path).get_attribute("title")
    print(co_cat)
    co_path_2 = f'//*[@id="listContents"]/div[2]/table/tbody/tr[{i}]/td[2]/span/a'
    co_name = driver.find_element(By.XPATH, co_path_2).text
    print(co_name)
    co_id = re.findall(r'\d{8}',driver.find_element(By.XPATH, co_path_2).get_attribute("href"))[0]
    print(co_id)

    # ==> 공시 이름, 공시 고유값
    # xpath : id --> 직접 가는 경로가 규칙화가 어려움!!!!
    # ==> full_xpath
    rcp_path = f'/html/body/div[4]/div[2]/div[1]/div[2]/div[3]/div[2]/table/tbody/tr[{i}]/td[3]/a'
    rcp_name = driver.find_element(By.XPATH, rcp_path).text
    print(rcp_name)
    rcp_id = re.findall(r'\d{14}',driver.find_element(By.XPATH, rcp_path).get_attribute("href"))[0]
    print(rcp_id)

    # 요청 일자는 skip --> 스스로 채워보기
    result.append( [t_data, co_cat,co_name,co_id, rcp_id,rcp_name ])
    print("-"*50)
print("Done!!!")
```
