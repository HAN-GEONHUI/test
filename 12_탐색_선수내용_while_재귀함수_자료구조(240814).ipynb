```python
# 예) 1부터 20까지 출력하자!!!
#     ===> for문 돌릴 20개 딱 만들어서 롤링!!
for i in range(1,21):
  print(i)
```

```python
#while 과점 --> 조건으로 바라보면..
#m1) 값을 늘리는 방향으로 설계!
i=1 # --> 출력을 위한 변수
while i<21: #이 조건만 만족하면... 주구장창 돌리자! 몇 번 관심 X
    print(i) #파이썬은 무한루프 방지 --> 중간에 세움!
    i +=1
```

```python
# 다른 설계 방식 : 값을 줄이는 스타일로 구현
i=20
while i>0:
  print(21-i)
  i-=1
```

```python
# m3) 명확하게 때려칠 조건을 작성해서 그만두기!!!
i = 1
while True: # 주구장창 무한 loop ==> 반복문을 강제로 세우는 break
    print(i)
    i+=1
    if i ==21:
        break
```

```python
# m3-1) 언제 세울지에 위치/구조에 따라서 코드는 다를 수 있음.
i = 1
while True:
    print(i)
    if i==20:
        break
    i+=1
```

```python
#선수 내용 1-2) pass, continue,break
#-pass: 아래 내용 skip --> 형식상 맞춤을 위해서 사용!
#       코테에서는 거의 pass 사용을 안하는 경우가 많아요
#- continue: for/while에서 다음 순번으로 넘길 때 사용!
#-break: 반복문 for/while을 그냥 다음 것들이 남아도 완전 세우는 역할!!
```

```python
#break 예
a = [3,4,5,6,7,8,9]
for i in a:
  print("체크 전의 값:",i)
  #중간에 혹시 5가 나온다면 뒤에 뭐가 있더라도 때려치자!
  if i ==5:
    print(i)
    break
```

```python
#pass 예
a = [3,4,5,6,7,8,9]
for i in a:
  print("체크 전의 값:",i)
  #중간에 혹시 5가 나온다면 뒤에 뭐가 있더라도 때려치자!
  if i ==5:
    print(i)
    pass
    print(i+10000)   #---> 요 값이 결과 값에 포함 됨
  print("passssssssssssssss")
```

```python
#continue 예
a = [3,4,5,6,7,8,9]
for i in a:
  print("체크 전의 값:",i)
  #중간에 혹시 5가 나온다면 뒤에 뭐가 있더라도 때려치자!
  if i ==5:
    print(i)
    continue
    print(i+10000)   #위에랑 다르게 결과값에 포함 안되고 지나감
  print("passsssssssss")
```

```python
#재귀 함수 예) 4!
#직접 계산하는 과정:4!=4*3*2*!

#1)for
n=4
fact =1
for i in range(1,n+1,1):
  fact *=i
print(fact)
```

```python
# 2) while 문
# ===> 조건이 만족할 때 까지 주구장창!!!!
#      편한 점 : 도는 횟수는 신경 안 써도 됨!!!!
#                도는 조건/환경 중심으로 작성!!!!
#      단점 : 종료 조건 설계!!!!
n = 4
fact = 1

cnt = 1 #<--- while 빠져나갈 기능 적인 세팅
while cnt <=n:
  fact *=cnt
  cnt +=1 #<--- 종료를 위한 모니터링 변수 조작
print(fact)
```

```python
#-->!할 일: 정의가 아니라 기능 중심 바로보자!
#   점화식
#   n!=n*(n-1)!,   1!=1
#-->"기능 중심"의 "규칙"!!!
#--> 바로 코드화
def my_fact(n):
  #호출이 되었는지 체크--> print
  print("MyFact호출?"+str(n))
  if n<=1:
    return 1
  else:
    return n*my_fact(n-1)
# ---> 기능적인 규칙성을 그냥 코드화 작업이 가능함!!!!
#      코드가 상당히 직관적으로 작성이 됨!!!!!
# ===> 규칙성이 있는 것들에 대한 관계 중심의 코드 작성에 용이함!!!
#      퀵정렬,탐색(dfs),dp,규칙적인 부분 구현: 코테용
```
