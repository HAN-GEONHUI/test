```python
from numpy import nan as NA
import pandas as pd
```

```python
data = pd.DataFrame(
    data = [
        [1,6.5, 3],
        [1, NA,NA],
        [NA,NA,NA],
        [NA, 6.5,3]
    ]
)
data
```

```python
data.dropna()
#--> 가로줄을 기준으로 모든 정보가 있는 경우를 살리는 것
#    (모든 컬럼에 대해서 nan이 아닌 경우에 대한 샘플만!)
#장점:샘플이 모든 속성 값을 가지고 있음..""빈 문자열 고민!
#단점:샘플의 수가 확 줄 수 있어요;;;;
```

```python
data = pd.DataFrame(
    data = [
        [1,6.5, 3],
        [1, NA,NA],
        [NA,NA,NA],
        [NA, 6.5,3]
    ]
)
data
```

```python
data.dropna(axis=0)
# --> 가로줄 기준으로 모든 세로값이 있는 것만..
```

```python
data.dropna(axis=1)
# --> 세로줄 기준으로 모든 가로의 값이 있는 것만 살리자!!!
```

```python
# dropna에서
# axis : 가로줄/세로줄로 데이터들이 다 있는지 체크!
#        axis=0/1
#thresh: 정상적인 데이터가 살아 있는 기준의 수!!
#how="all": 그 줄에 모든 값이 누란된 경우만 제거해주세요!
```

```python
data.dropna(axis=0)
# --> 가로줄 기준으로 모든 세로값이 있는 것만..
```

```python
data.dropna(axis=0,how="all")
```

```python
data.dropna(axis=0,thresh=1)
#가로줄 기준으로 모든 칼럼의 정상적인 값이 1개 이상이면 살리자!
```

```python
data.dropna(axis=0, thresh=2)
# --> 가로줄 기준으로 모든 컬럼의 정상적인 값이 2개 이상이면 살리자!
```

```python
# 채우자 fillna ==> nan로 되어 있는 값을 채우자!--> 컬럼별로 누락된 값을 각각 채우고 싶다!!
# 예) 1번 컬럼의 누락데이터는 기본값을 0으로 채우고,2번 컬럼의 누락데이터는 기본값을 9999로 채우고 싶다
# 참고) 여기서는 기능적인 부분이여서 특정한 값으로 채우는 경우만 할 예정
```

```python
data.fillna(100)
```

```python
data.fillna({0:0,1:999})
#--> 어느 컬럼은 뭐로 누락된 값을 채우고, 어느 컬럼은 무엇으로 누락값을 채울지
#--> dict 쌍 구조
```
