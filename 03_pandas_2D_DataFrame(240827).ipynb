```python
# To Do)
# 위의 DF에서 새로운 GapFor 컬럼에는 apply+lambda말고 그냥 날짜별로 돌아가면서 for문을 사용해서
# 그 날짜의 시가, 종가, 고가, 저가의 값들을 바탕으로 최고값을 찾고, 최저값을 찾아서, 그 차이를
# GapFor 컬럼에 기록하기

# ==> 구조!!전체 DF을 batch!!!!
# 직접 해보기!! for문 이용해서!! DF의 값의 접근을 이용해서 DF의 값의 접근을 이용하기
#앞에서 한 pandas 잡다한 기능을 &체크
#for: 가로줄 index(날짜), 정수 인덱스
```

```python
for i in range(len(stock_pirce_df)): #가로줄을 돌려준다? 무슨소리지
  print(i)
```

```python
#기본적인 접근 방식!
#--> 모든 "가로줄"을 돌아가면서 : for/apply : for이 가로줄을 돌아가면서 출력해주는데 apply도 똑같다.
#--> i번째 가로줄에 무엇을 할지!
#     1) 가로 i번째,세로: 시종고저(0~3): 4개 값--> loc/iloc
#     2) 1번의 4개의 값 중에서 :max
#     3) 1번의 4개의 값 중에서 :min
#     4) max_min 계산
#     5) 4번의 계산 값을 어디에 기록할지
#             가로:i번째,세로: GapFor위치--> at/iat
```

```python
# sol1) 태생적인 정수 인덱스 중심으로 하자!!!!!
# ------ 참고 : stock_pirce_df.shape[0], len()

for i in range( len(stock_pirce_df)):
    # i번쨰 가로줄에서 필요한 4개 값 가져오기!!!!!
    data = stock_pirce_df.iloc[i, :4] # --> Series
    # max-min :계산
    temp = data.max()-data.min()
    #i번째 가로줄,세로 GapFor위치 : 5번째
    stock_pirce_df.iat[i,5] = temp
stock_pirce_df
```

```python
#sol1 수정!!) 정수로 세팅!:get_loc
cols = ["시가","고가","저가","종가"]
#--> 필요한 컬럼들을 리스트업
#--> 이에 대응하는 컬럼의 정수 인덱스
cols_idx = [stock_pirce_df.columns.get_loc(col)for col in cols]

new_col_name = "GapFor2"
stock_pirce_df[new_col_name]=0
gap2_idx = stock_pirce_df.columns.get_loc(new_col_name)

for i in range( len(stock_pirce_df)):
    # i번쨰 가로줄에서 필요한 4개 값 가져오기!
    data = stock_pirce_df.iloc[i, cols_idx] # --> Series
    # max-min : 계산
    temp = data.max() - data.min()
    # i번째 가로줄, 세로 GapFor위치 : 5번째
    stock_pirce_df.iat[i,gap2_idx] = temp
stock_pirce_df
```

```python
# sol2) 내가 만든 인덱스로 돌리자!!
# ==> 날짜 중심으로 돌립시자!!
cols = ["시가","종가","고가","저가"]
result_col ="GapFor3"
stock_pirce_df[result_col] = 0

for i in stock_pirce_df.index:
    # i번째 내가 만든것들 중심의 가로줄!!!
    data = stock_pirce_df.loc[ i, cols]
    temp = data.max()- data.min()
    stock_pirce_df.at[i, result_col] = temp
stock_pirce_df
```
