```python
# 요청하려는 api의 주소 작성...
# 1) base url
base_url =" https://kobis.or.kr/kobisopenapi/webservice/rest/movie/searchMovieList.xml"

key = "2526481d2469f017266060b5a7cdff1f"
url_s1= "40"
url = base_url+"?key="+key +"&itemPerPage="+url_s1


url
```

```python
# Step2) 위에서 작성한 주소를 통신으로 요청을 받아서
#        ==> 받은 후 BS활용해서 Tag 바로 찾을 수있도록 하기
res = urllib.request.urlopen(url)
#->xml양식으로 요청한 정보들을 처리하기 위해서는 bs->parser:"html.paser","lxml.parser"etc,,
#     보통 일반적인 경우들에 있어서는 html.parser를 사용해도 무방하고 xml의 사이즈가 크거나 하면
#별도의 전문 큰사이즈 파서 사용하고 메뉴얼을 참조하면 됨!!!!
soup = BeautifulSoup( res, "html.parser")
```

```python
#for 문을 총해서 돌리기 위해서
#--> 위의 처리하다 발생하는 에러를 보고 에러 처리를 추가해 보겠습니다.



tot_cnt = len(soup.find_all("movie"))

for idx in range(tot_cnt):
  i_movie = soup.find_all("movie")[idx]
  print(i_movie.find("moviecd").text)
  print(i_movie.find("movienm").text)
  print(i_movie.find("movienmen").text)
  print(i_movie.find("opendt").text)
  if i_movie.find_all("peoplenm") !=[]:
    print(i_movie.find("peoplenm").text)
  else:
    print("감독X")
  print("*"*100)
```

```python
# Q) 어제 한대로 수집한 영화 정보들에 대해서 DF에 담아보기
#    ==> list 기반으로도 해보고, dict 기반으로 해보기
```

```python
movie_list = []
for idx, data in enumerate( soup.find_all("movie")):
    #  data : 개별 영화에 대한  tag 작성된 정보!!!
    # ** 중요 ** 데이터 정보에 대한 추출의 단위 : 영화/샘플!!
    code = data.find("moviecd").text
    name = data.find("movienm").text
    ename = data.find("movienmen").text
    date = data.find("opendt").text
    if data.find_all("peoplenm") != []:
        dir_name = data.find("peoplenm").text
    else:
        dir_name ="X" # --> 빼면 밀릴 수 있으니 주의!!!
    movie_list.append( [code,name,ename, date, dir_name ])
print("Done!!")
```

```python
movie_df=pd.DataFrame(movie_list,
                      columns=["Code","Title",
                               "E-Title","openday",
                               "Dir_name"])
movie_df.head()
```

```python
movie_list = []
for idx, data in enumerate( soup.find_all("movie")):
    #  data : 개별 영화에 대한  tag 작성된 정보!!!
    # ** 중요 ** 데이터 정보에 대한 추출의 단위 : 영화/샘플!!
    i_dict = {"code":"", "title":"", "e-title":"",
              "openday":"", "dir_name":"X"}
    i_dict["code"] = data.find("moviecd").text
    i_dict["title"]= data.find("movienm").text
    i_dict["e-title"]= data.find("movienmen").text
    i_dict["openday"]= data.find("opendt").text
    if data.find_all("peoplenm") != []:
        i_dict["dir_name"] = data.find("peoplenm").text
    movie_list.append( i_dict )
print("Done!!")
```
