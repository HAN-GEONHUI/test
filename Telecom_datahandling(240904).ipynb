```python
#Q)이탈하지 않은 고객들을 대상으로(Churn=0): 잘 사용하는 고객들을 대상으로 국제 여행 계획이 있는 분드를 대상 중에서 Internation Plan =Y/N
#--> 전체 국제 통화 요금 중 최고값은 얼마?
#    total international charge
# 의도) 필터링
```

```python
# 1) 가로줄에 대한 필터링 조건 : 2개
# 1-1) Churn ==0
# 1-2) Internaiotnal Plan = Yes
# ==> 위의 2개의 조건의 관계 : and

# 2) 세로줄에 대한 필터링/보고자하는 항목/속성 : 국제 통화 요금
#df[조건]
#df[조건1 (and,or) 조건2,,,,,]
#df.loc[ 가로 필터링에 대한 조건들, 볼 컬럼들에 대해서 선택]
```

```python
#data["Churn"]==0
#data["International plan"]=="Yes"
data.loc[(data["Churn"]==0) & (data["International plan"]=="Yes"), "Total intl charge"]
```

```python
data.loc[(data["Churn"]==0) & (data["International plan"]=="Yes"), "Total intl charge"].max()
```

```python
# cf) 이탈하지 않고, 국제 계획이 없는 고객들 중에서
#     가장 많은 국제요금을 지불한 금액은 얼마?
data.loc[(data["Churn"]==0) & (data["International plan"]=="No"), "Total intl charge"].max()
```

```python
#Q) State 컬럼에는 51개의 모든 주에 대한 정보들이 있음-> 1)기존의 값을 재조정 하고 싶다 2)간단하게 주의 코드의 앞 대문자 가지고 재정의를 하겠다
#-> 수집한 값을 원하는 대로 변경하고자 한다
#f-code 라는 새로운 컬럼에 기존 area_code의 값으로 재변경, f-code의 값이 "W"인 데이터만 추려보기(몇개 나오는지)
```

```python
data["State"].value_counts()
```

```python
data["f-code"]=data["State"].apply(lambda x: x[0])
```

```python
data.head()
```

```python
#data.loc[data.loc["f-code",:]=="W",:]
data[data["f-code"]=="W"]
```

```python
# + 참고) 위에와 다르게 굳이 컬럼을 만들지 않고 단순하게 체크/조회만 하려고 한다면
# ==> 기존에는 주어진 값을 기준으로 필터링 조건을 작성
#     ++ 필터링 조건에 기존의 값이 아니라 내가 변경하고자 하는 기준으로 필터링!!
data.loc[data.loc[:,"State"].apply(lambda x: x[0]) == "W",:]
```

```python
# *** 중간에 데이터를 탐색을 할 때 ===> 주어진 값/수집한 값 그자체로 보는 것이 아니라 내가 생각하거나,
#                                       내가 생각하는 가성,도메인 지식을 활용해서 새로운 기준에 대한 값으로 필터링 하려고 하면
#-->불리언 인덱싱 + apply&lambda
```
