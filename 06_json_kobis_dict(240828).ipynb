```python
#필요한 모듈
import pandas as pd
import requests  #--> colab,anaconda 기본 패키지
#import josn 생략 가능함
```

```python
# 요청하려는 api의 주소 작성...
# 1) base url
base_url =" http://www.kobis.or.kr/kobisopenapi/webservice/rest/movie/searchMovieList.json"
# 2) 실제 요청 파라미터에 대해서 dict로 표현!!!
#    api상에서 사용할 수 있는 요청인터페이스 : key(문자열)
#                             요청인자에 대한 값 : value(숫자, 문자.)
# --> key, itemPerPage
my_params = {
    "key" : "2526481d2469f017266060b5a7cdff1f",
    "itemPerPage" : 40
}
# 3) 해당하는 요청에 대한
res = requests.get(base_url, params=my_params)

# 4) 중간 FM으로 http 통신에 요청을 하면,,,http 응답 코드 200--> OK
#    https://developer.mozilla.org/en-US/docs/Web/HTTP/Status#successful_responses
if res.ok:
    result = res.json()
else:
    print("Check your HTTP Response Code!!!")
result
```

```python
# +++ Dict 통해서 감독에 대한 정보를 처리하기 위해서..
# ==> 1번 감독 이름에 대한 리스트업..
directors_list = []
for idx, data in enumerate(result["movieListResult"]["movieList"] ):
    directors_list.append( data["directors"])
directors_list
```

```python
#기호 본 경험
#1) *args : 파이썬sum_random(*args):--> 리스트
#2)**args : def aa(p="12",cc=123)
#    dict
```

```python
# --> 전체 영화에 대한 정보들을 담을 리스트...
new_list = []
# idx : 정수 인덱스 : 감독 정보 처리...
# data : 개별 영화에 대한 정보 : dict
for idx, data in enumerate(result["movieListResult"]["movieList"] ):
    one_data = data
    if len(directors_list[idx]) != 0:
        one_data = {**one_data, **directors_list[idx][0]}
        # {"movieCd":1234, "movieNm":"광해"}
        # {"peopleNm":"AAA"}
        # {"movieCd":1234, "movieNm":"광해","peopleNm":"AAA"}
    new_list.append( one_data)
print(new_list)
```
