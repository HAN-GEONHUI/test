```python
#단편 영화에 맞는 typeNm값으로 교체함.
```

```python
url_p1 = "http://www.kobis.or.kr/kobisopenapi/webservice/rest/movie/searchMovieList.json"
key = "2526481d2469f017266060b5a7cdff1f"
url_p2 = "40"
type_nm = "220102"  # 단편 영화에 맞는 typeNm 값으로 교체
url = url_p1 + "?key=" + key +"&typeNm=" + type_nm + "&itemPerPage=" + url_p2  + "&movieTypeCd=" + type_nm
print(url)
```

```python
movie_page = urllib.request.urlopen(url)
movie_data = json.loads( movie_page.read().decode("utf-8"))
movie_data
```

```python
#코드/이름/감독이름/배우/장르/상영시간
#movieCd/movieTitle/movieETitle/directors/DirName/repGenreNm/showTm
```

```python

tot_data = []
for idx,data in enumerate(movie_data["movieListResult"]["movieList"]):
  i_code = data["movieCd"]
  i_name = data["movieNm"]
  i_name_e = data["movieNmEn"]
  i_gen = data["repGenreNm"]

  if data["directors"] != []:
    i_dir=""

  if data["actors"] != []:
    i_act= ""

  if data["showTm"] !=[]:
    i_tm =""

    tot_data.append([i_code,i_name,i_name_e,i_act,i_gen,i_tm,i_dir])
print('Done!!')
```

```python

tot_data = []
for idx,data in enumerate(movie_data["movieListResult"]["movieList"]):
  i_code = data["movieCd"]
  i_name = data["movieNm"]
  i_name_e = data["movieNmEn"]
  i_gen = data["repGenreNm"]

  if data["directors"] != []:
    i_dir = data["directors"][0]["peopleNm"]
  else:
    i_dir=""




    tot_data.append([i_code,i_name,i_name_e,i_gen,i_dir])
print('Done!!')
```

```python
tot_data
```

```python
movie_df = pd.DataFrame(
    data = tot_data,
    columns=["코드","이름","영어이름","장르","감독이름"]
)
movie_df
```

```python
movie_df.set_index( "코드")
```
