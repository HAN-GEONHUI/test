```python
import time
import timeit
#--> 시간과 관련된 부분 측정
```

```python
#데이터를 합칠 때 주로 사용되는 자료형
#-->리스트/dict + deque
#   numpy array
#     pandas DF
import pandas as pd
import numpy as np
from collections import deque
```

```python
#입력으로 추가할 샘플의 수:n
```

```python
# DF을 기준으로
# 1개의 DF을 전체 DF에 하나씩 쌓아가는 방식==> DF으로 시작해서, DF에 쌓자!
def df_concat(n):
    df = pd.DataFrame(columns=["A","B","C"])
    temp = pd.DataFrame.from_dict({"A":[1], "B":[11.3], "C":["abcdef"]})
    for _ in range(n):
        df = pd.concat([df, temp], ignore_index=True)
    return df
```

```python
#개별 데이터를 파이썬의 리스트로 핸들링-->1개의 샘플이 수집이 될 때마다 리스트에 쌓아두고 다 수집하면 1번 최종적으로 DF으로 변환
def list_append_df(n):
    data = []
    for _ in range(n):
        data.append([1,11.3,"abcdef"])
    return pd.DataFrame(data, columns=["A","B","C"])
    # [
#        [ 1,11.3,"ㅁㅍ"],
#        [1,11.3,"ㅇㅁ"]
    # ]
```

```python
# numpy의 array를 대상으로 모든 것을 처리!==> 다 데이터를 수집을 하고, 최종적인  DF
def np_append_df(n):
    data = np.array(np.empty((0,3), int))
    for _ in range(n):
        data = np.append(data, np.array([[1, 11.3,"abcdef"]]), axis=0)
    return pd.DataFrame(data, columns=["A","B","C"])
```

```python
def deque_append_df(n):
    data = deque([])
    for _ in range(n):
        data.append([1,11.3,"abcdef"])
    return pd.DataFrame(data, columns=["A","B","C"])
```

```python
# Test를 잠시 수행!!
# --> 수집할 데이터 적은 경우 : 거의 사람이 못 느낄 정도..
n = 10
%time df_concat(n)
%time list_append_df(n)
%time np_append_df(n)
%time deque_append_df(n)
```

```python
n = 10000
%time df_concat(n)
%time list_append_df(n)
%time np_append_df(n)
%time deque_append_df(n)
```
