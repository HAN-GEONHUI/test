```python
# 1) 시간 변환 함수
# 입력) HH:MM
# 할일) 입력HH:MM을 00:00기준으로 몇 분이 지났는지 분으로 계산
# 출력) 지난간 분 시간
```

```python
def convert_h_to_m(time_HHMM):
  h_part,m_part = time_HHMM.split(":")
  total_min = int(h_part) * 60+ int(m_part) # 산수+위해서 형변형
  return total_min
```

```python
# 본 코드 작성
def solution(fees, records):
    answer = []
    #####################
    # 1) feees의 요금 관련 정보 처리 --> 기본 시간, 기본 요금, 기준시간, 기준요금
    # ==> 정수 인덱스..:멀티 할당-> 변수화
    base_time, base_cost, add_time, add_cost = fees
    # 참고) 변수 할당 안 하고, fees[0], fees[1] ,,, :직접 사용해도 상관 없음!!!

    # 2) 입출에 대한 정보를 처리!!!!--> 마라톤//신고자 유사...
    # ==> 처리를 Dict 자료형을 정리!!!
    #     key : 자동차 번호
    #     value : 차량의 입출 정보들--> (HH:MM을 시간을 M분으로 처리, 상태)
    # ==> 기존에 키로 등록된 차량인지, 아닌지에 따라서 처리가 조금 다름!
    # 예) { 5961 : [ (5*60+34, IN), (7*60+32, OUT) ,,,,,,,,]}
    park_dict = {}
    # 입력에 대한 주차 로그 기록지를 보면서..정리...
    for record in records: # ["01:23 0000 IN", "03:45 1111 IN",,,]
        # 2-1 1개 기록처리) 시간, 차번호, 입출상태 : 공백구분!!
        park_time, car_id, status = record.split(" ")
        # ["01:23", "0000", "IN"]
        # 2-2 정리를 차 번호 기준 k) key로 사용 "0000"--> int---> 0
        car_id = int(car_id) # key
        # 2-3 정리를 차 번호 기준에 상태 정보 v)
        # (시간, 상태) ["01:23", "IN"] --> 시간변환함수 -->[1*60+23, "IN"]
        park_time, car_id, status = record.split(" ")
        car_record = [convert_h_to_m(park_time), status ] # value
        # 2-4) 실제 등록 처리!!!!!!
        # case1) 이미 등록된 차 번호
        if car_id in park_dict:
            park_dict[car_id].append( car_record ) # [90,"OUT"]
            # {0:[ [83,"IN"], [90,"OUT"]]}
        # case2) 신규 등록할 차 번호
        else:
            # {0:[[83,"IN"]]}
            park_dict[car_id] = [car_record]# [83,"IN"]  --> [ [83,IN] ]
            # value의 타입이 리스트 -->  append 추가 기록 추가할 수 있음.

    # 3) 누적 시간 + 비용 계산!!!!
    # ==> 계산용에 대한 dict ( key : 차번, value : 전체 요금) : 출력용
    # Check1) 출차 정보 처리: 출차 정보가 없으면 추가!!!
    # Check2) 누적 시간 계산
    #          2-1) 기본 요금 이하
    #          2-2) 기본 요금 초과
    #               2-2-1) 초과 시간이 나눠 떨어질 떄 : 몫 그래도 사용
    #               2-2-2) 초과 시간이 나눠 떨어지지 않으면 : 몫+1올림
    car_dict = {} # 키: 차번호, 값: 요금
    for car, logs in park_dict.items(): # k,v
        #  개별 자동차를 기준으로 처리가 이루어짐!!
        # logs = [ [83,IN], [90,out], [100, in]]
        # ==> in : [83, 100]
        # --> out : [90]
        out_time_list = [ m for m,sta in logs if sta=="OUT"]
        in_time_list  = [m for m,sta in logs if sta=="IN"]
        # 들어왔는데,,,,나간 정보가 없는 차량들 처리 : logs 갯수 홀수
        if len(logs) % 2 ==1: # 출차 정보가 없는 경우
            out_time_list.append(convert_h_to_m("23:59") )
        #===> 출차 정보가 없는 경우를 처리!!!!!!!

        # 차량별로 누적 시간 계산 : 분
        total_min = sum(out_time_list) - sum(in_time_list)

        # 요금 계산 수행!!!!
        if total_min <= base_time: # fees[0]
            # 기본 시간 이내 누적 주차 시간 : 기본 요금
            car_dict[car] = base_cost # fees[1]
        else: # 추가 요금이 발생!!!!
            # 나눠 떨어질 떄와 아닐 떄를 구별해서 진행!!!!
            # if 한 번 안 사용하기 위해서 -부호 // 연산자!!!
            cost = 0
            cost += base_cost
            total_min -= base_time # 누적시간 = 기존누적시간 - 기본 시간
            # 올림에 대한 처리 : 기술적인//.. if 문으로 하셔도 됨!!!
            over_time_base = -( total_min* (-1) // add_time )
            cost += over_time_base * add_cost
            # 최종 추가 시간있는 차량의 요금
            car_dict[car] = cost
    # --> car_dict = { 0: 14600, 148:34400,5961:5000}

    # 4) 최종 제출 답안지 작성!!!
    #   ==> 정렬 : key에 차량 번호 오름차순
    #   ==> 출력 정보 : 오로지 요금만 : values--> 리스트 양식으로...
    # 참고) dict/ list 왔다 갔다하는 것은 필수적인 것이 아니라 기능적 소개!!!
    car_cost = dict(sorted(car_dict.items()))
    answer = list(car_cost.values())
    # ===> 형 변화에 대한 것들을 자유롭게 하실 수 있도록 유연하게 생각!!!

    ####################
    return answer
```

```python
# ==> 코드 정리

def convert_h_to_m( time_HHMM):
    h_part, m_part = time_HHMM.split(":")
    total_min = int(h_part) * 60 + int(m_part) # 산수+위해서 형변형
    return total_min

def solution(fees, records):
    answer = []
    base_time, base_cost, add_time, add_cost = fees
    park_dict = {}
    for record in records: # ["01:23 0000 IN", "03:45 1111 IN",,,]
        park_time, car_id, status = record.split(" ")
        car_id = int(car_id) # key
        park_time, car_id, status = record.split(" ")
        car_record = [convert_h_to_m(park_time), status ] # value
        if car_id in park_dict:
            park_dict[car_id].append( car_record ) # [90,"OUT"]
        else:
            park_dict[car_id] = [car_record]# [83,"IN"]  --> [ [83,IN] ]
    car_dict = {} # 키: 차번호, 값: 요금
    for car, logs in park_dict.items(): # k,v
        out_time_list = [ m for m,sta in logs if sta=="OUT"]
        in_time_list  = [m for m,sta in logs if sta=="IN"]
        if len(logs) % 2 ==1: # 출차 정보가 없는 경우
            out_time_list.append(convert_h_to_m("23:59") )
        total_min = sum(out_time_list) - sum(in_time_list)
        if total_min <= base_time: # fees[0]
            car_dict[car] = base_cost # fees[1]
        else: # 추가 요금이 발생!!!!
            cost = 0
            cost += base_cost
            total_min -= base_time # 누적시간 = 기존누적시간 - 기본 시간
            over_time_base = -( total_min* (-1) // add_time )
            cost += over_time_base * add_cost
            car_dict[car] = cost
    car_cost = dict(sorted(car_dict.items()))
    answer = list(car_cost.values())
    return answer
```
