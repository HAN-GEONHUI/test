```python
# 본인 환경에 맞는 엑셀 파일 경로 설정
path = '/07_surveys_excelformat.xls'
data = pd.read_excel(path, sheet_name ="Sheet1")
# ==> 실제 엑셀 파일에서 열어 보시고,,,sheet 이름을 대소문자/공백들을 주의해서 지정하면 됨!!
data.head()
```

```python
# 받은 파일을 대상으로,,열심히 전처리를 하고,,
# 할 일이 다 끝났다고 한다면,,
# ==> DF을 출력을 하려고 함..
# csv파일
# xls(구번전), xlsx(신버전)
# df.to_excel(경로~~~)
```

```python
data.to_excel("/content/test.xlsx")
```
